```python
def fromDecimalToBase(base, number):
    newNumber = ""
    letters = 'ABCDEF'
    if number == 0:
        return "0"
    while number:
        if number % base > 9:
            newNumber = letters[(number % base) - 10] + newNumber
        else:
            newNumber = str(int(number % base)) + newNumber

        number //= base
    return newNumber
```
